```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_pb = pd.read_csv('../result/all_amino_salt_pb.csv')
df_gb = pd.read_csv('../result/all_amino_single_core.csv')
df_opt = pd.read_csv('../result/all_amino_with_optradii.csv')
# Remove the '_1' suffix from mutations in df_pb
df_pb['mutations'] = df_pb['mutations'].str.replace('_1', '')

# Perform an inner join on the mutations column
df_gb = pd.merge(df_gb, df_pb, on='mutations', how='inner')

pb_err = df_pb['delta_delta_g_pb_error']
gb_err = df_gb['delta_delta_g_gen_66_error_y']
```

```python
# Create the plot
plt.figure(figsize=(16, 12))
plt.plot(pb_err, 'o', label='PB', markersize=12)
plt.plot(gb_err, 'o', label='GBNSR6', markersize=12)

# Change x-axis labels to the mutations column in df_gb
plt.xticks(ticks=range(len(df_gb['mutations'])), labels=df_gb['mutations'], rotation=30, fontsize=25)
plt.yticks(fontsize=20)
plt.xlabel('Mutations', fontsize=20)
plt.ylabel('Error', fontsize=20)

# Add legend and show plot
plt.legend(fontsize=20)
# plt.savefig('~/Downloads/pb_gb.png', bbox_inches='tight')
plt.show()
```
